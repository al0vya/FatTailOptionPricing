```python
import matplotlib.pyplot as plt
import matplotlib.pylab  as pylab  

import numpy as np
import scipy.stats

from mpl_toolkits.mplot3d import Axes3D

params = {
    "legend.fontsize" : "xx-large",
    "axes.labelsize"  : "xx-large",
    "axes.titlesize"  : "xx-large",
    "xtick.labelsize" : "xx-large",
    "ytick.labelsize" : "xx-large"
}

pylab.rcParams.update(params)
```

```python
class PDFGaussianFolded(
        scipy.stats.rv_continuous
    ):
        def _pdf(
                self,
                x
            ):
                # have a variance equal to that of the invented distribution
                sigma = np.pi * np.pi / 12
                
                return 1 / sigma / np.sqrt(2 * np.pi) * np.exp( -0.5 * (x / sigma) * (x / sigma) )
```

```python
class PDFInventedFolded(
        scipy.stats.rv_continuous
    ):
        def _pdf(
                self,
                x
            ):
                return 0.5 * ( 1 / np.cosh(x) ) * ( 1 / np.cosh(x) )
```

```python
def compare_invented_and_gaussian():
    x = np.linspace(0, 4, 100)
    
    P_invented = PDFInventedFolded(momtype=0, a=0).pdf(x)
    P_Gauss    = PDFGaussianFolded(momtype=0, a=0).pdf(x)
    
    plt.plot(x, P_invented, label="Invented")
    plt.plot(x, P_Gauss,    label="Gaussian")
    plt.ylabel("Probability")
    plt.xlabel('x')
    plt.legend()
    plt.ylim(0, 0.55)
    plt.xlim(0, 4)

compare_invented_and_gaussian()
```

```python
def survival_invented_vs_gaussian():
    x = np.linspace(0, 1, 100)
    
    Fbar_invented = PDFInventedFolded(momtype=0, a=0).sf(x)
    Fbar_Gauss    = PDFGaussianFolded(momtype=0, a=0).sf(x)
    
    plt.plot(x, Fbar_invented, label="Invented")
    plt.plot(x, Fbar_Gauss,    label="Gaussian")
    plt.ylabel("log( P(X > x) )")
    plt.xlabel("log(x)")
    plt.yscale("log")
    plt.xscale("log")
    plt.legend()
    
survival_invented_vs_gaussian()
```

```python
def survival_power_laws():
    """
    A subexponential distribution has the following limit definition:
    
         As x -> inf, ( 1 - convolution( F(x), F(x) ) ) / 1 - F(x) = 2
    
    Where F(x) is the cumulative probability distribution of x.
    
    Cumulative implies
    
        F(x) = P(X < x), 
        
    the probability the X we sample is less than x.
    
    The survival function,
    
        S(x) = 1 - F(x) = P(X > x),
        
    is the probability the X we sample is more than x.
    
    convolution( F(x), F(x) ) essentially means sample from F(x) twice.
    
    So, 1 - convolution( F(x), F(x) ) means probability we sample from F(x) twice and both times exceed x.
    
    When sampling X under these distributions, we now interpret the limit as:
    
        probability sample from F(x) twice and both times exceed x
        ---------------------------------------------------------- = 2
                      probability that either exceed x
    
    as x becomes bigger. As x becomes bigger, the more we are in the tails.
    
    Only in the tails does the limit hold.
    
    The limit may be surprising as we may expect that when we sample in the tails twice,
    the probability that we exceed x is more than double than just sampling once.
    
    However, the X from sampling once may put us close to the Xs from sampling twice as large values
    are "in a probabilistic sense close together" in the tails of subexponential distributions.
    """
    xmin = 0
    xmax = 10000
    
    x = np.linspace(xmin, xmax, 500)
    
    Fbar_pareto_2 = scipy.stats.pareto.sf(x, b=2-1, loc=-1)
    Fbar_pareto_3 = scipy.stats.pareto.sf(x, b=3-1, loc=-1)
    Fbar_pareto_4 = scipy.stats.pareto.sf(x, b=4-1, loc=-1)
    
    plt.plot(x, Fbar_pareto_2, label="a = 2")
    plt.plot(x, Fbar_pareto_3, label="a = 3")
    plt.plot(x, Fbar_pareto_4, label="a = 4")
    plt.ylabel("log( P(X > x) )")
    plt.xlabel("log(x)")
    plt.yscale("log")
    plt.xscale("log")
    plt.legend()
    
survival_power_laws()
```
